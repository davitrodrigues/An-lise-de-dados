h) Há mais aluguéis quando venta muito ou não? (Criar o campo venta_muito onde Sim será quando a velocidade do vento é maior que a média, Não caso o contrário)

```python
media_vento=df['windspeed'].mean()
media_vento
```

```python
def vento(x):
    if x>media_vento:
        return 'SIM'
    else:
        return 'NÃO'
```

```python
df['venta muito']=df['windspeed'].apply(vento)
df
```

```python
df.groupby('venta muito')['count'].sum()
```
