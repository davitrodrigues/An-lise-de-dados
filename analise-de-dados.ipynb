h) HÃ¡ mais aluguÃ©is quando venta muito ou nÃ£o? (Criar o campo venta_muito onde Sim serÃ¡ quando a velocidade do vento Ã© maior que a mÃ©dia, NÃ£o caso o contrÃ¡rio)

```python
media_vento=df['windspeed'].mean()
media_vento
```

```python
def vento(x):
    if x>media_vento:
        return 'SIM'
    else:
        return 'NÃƒO'
```

```python
df['venta muito']=df['windspeed'].apply(vento)
df
```

```python
df.groupby('venta muito')['count'].sum()
```
