Exercício 3 – Histograma (População)
Objetivo: Visualizar a distribuição da população.

Construa um histograma da população dos estados.

Defina um número apropriado de intervalos (bins).

Observe se a distribuição é simétrica, assimétrica, concentrada em algum intervalo, etc.

Repita o histograma com densidade (usando density=True), para comparar.



```python
limite=[0,5,10,15,20,25,30,35,40]
```

```python
ax = (df['Population'] / 1_000_000).plot.hist(bins=limite,figsize=(6,4))
ax.set_xlabel('População  (milhões)')
plt.tight_layout()
plt.show()
#sintaxe para historiagrama
#bin seriam os intervalos de cada coluna
```

```python
ax = (df['Population'] / 1_000_000).plot.hist(bins=limite,density=True,figsize=(6,4))
ax.set_xlabel('População  (milhões)')
plt.tight_layout()
plt.show()
```

```python
ax = (df['Population'] / 1_000_000).plot.hist(bins=limite,density=True,figsize=(6,4))
ax.set_xlabel('População  (milhões)')
plt.tight_layout()
plt.show()
```

A diferença para um que tem densidade e outro que não tem, é o que tem densidade mostra a probilidade para cada barra
